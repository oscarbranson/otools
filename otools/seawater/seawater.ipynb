```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
from otools.chemistry import seawater
```

```python
sw = pd.read_csv('seawater.csv', comment='#')

fig, ax = plt.subplots(figsize=(10, 4))

xlabels = []
for i, r in sw.iterrows():
    xlabels.append(r['name'])
    ax.plot([i,i], r[['min', 'max']])
    ax.scatter(i, r['value'], color='k', marker='_', s=10, zorder=2, alpha=0.6)
    
ax.set_yscale('log')
ax.set_xticks(range(len(xlabels)))
ax.set_xticklabels(xlabels, rotation=90, fontsize=8)

for i in [1e-15, 1e-12, 1e-9, 1e-6, 1e-3, 1]:
    ax.axhline(i, color='gray', linestyle='--', zorder=-2, lw=1)
ax.set_yticks([1e-15, 1e-12, 1e-9, 1e-6, 1e-3, 1])

ax.set_xlim(-1, len(xlabels))
```

```python
sw.set_index('name', inplace=True)
```

```python

osw = seawater()

for s in osw:
    if s == 'SO4':
        os = 'S'
    else:
        os = s
    print(s, osw[s], sw.loc[os, 'value'], osw[s] / sw.loc[os, 'value'])
```
